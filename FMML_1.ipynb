```python
import numpy as np
import matplotlib.pyplot as plot
#Transpose
mat1=[[1,2,3],[1,1,1],[2,3,4]]
mat2=np.transpose(mat1)
print(mat1)
print(mat2)
#Dot product
scalar=np.dot(mat2,mat1)
print("\nDot Product\n")
print(scalar)
#Numpy Array
arr1=np.array(mat1)
arr2=np.array(mat2)
result=arr1.dot(arr2)
print("\n Dot product using Numpy Array \n")
print(result)
#Multi Dot Product
m1=np.random.randint(1,10,(5,10))
m2=np.random.randint(2,10,(10,15))
m3=np.random.randint(3,10,(15,20))
m4=np.random.randint(4,10,(20,25))
v=np.linalg.multi_dot([m1,m2,m3,m4])
print("\nMulti Dot Product\n")
print(v)
#Inner Product
inner=np.inner(arr1,arr2)
print("\nInner Product\n")
print(inner)
#Outer Product
outer=np.outer(arr1,arr2)
print("\nOuter Product\n")
print(outer)
#Multiplication using @
val=arr1@arr2
print("\nMultilpication using @\n")
print(val)
#Determinant
md1=np.random.randint(1,10,(3,3))
det=np.linalg.det(md1)
print("\nDeterminant\n")
print(det)
#Inverse
inv=np.linalg.inv(md1)
print("\nOriginal\n")
print(md1)
print("\nInverse\n")
print(inv)
#Matrix Powers
mp=np.random.random([5,5])
mp3=np.linalg.matrix_power(mp,3)
print("\nmatrix to the power 3\n")
print(mp3)
mp4inv=np.linalg.matrix_power(mp,-4)
print("\nMatrix inverse to the power 4\n")
print(mp4inv)
#Eigen Vectors and Eigen Values
eig_val,eig_vec=np.linalg.eig(mp)
print("\n Eigen Values and Eigen Values are \n")
print(eig_val)
print(eig_vec)
#Trace
mpt=np.eye(9)
print(mpt)
trace=np.trace(mpt)
print("\n Trace of matrix is \n")
print(trace)
#Norm
norm=np.linalg.norm(mpt)
print("\n Norm \n")
print(norm)
#Solving System of equations
soe=np.random.randint(1,100,[3,3])
dep=[1,90,34]
sol=np.linalg.solve(soe,dep)
print("\n Coefficient Matrix \n")
print(soe)
print("\n Dependent Matrix \n")
print(dep)
print("\n Solution \n")
print(sol)
```

```python
from skimage import data
import matplotlib.pyplot as plt

image = data.coffee()
plt.imshow(image)
print(image.dtype,image.min(),image.max(),image.shape)
plt.title('Coffee')
plt.show()
img=image/255.0
img=np.transpose(img,axes=(2,0,1))
u,s,v=np.linalg.svd(img)
print(u.shape,s.shape,v.shape)
#creating the sigma matrix
sigma = np.zeros((3,400,600))
for i in range(3):
  np.fill_diagonal(sigma[i,:,:], s[i,:])
print(sigma.shape)
#construction of matrix with any dimensionality reduction
construct = u @ sigma @ v
construct = np.transpose(construct, axes=(1,2,0))
plt.imshow(construct)
plt.show()
#construction using only 10 significant data points
k = 10
reconst = u @ sigma[:,:,:k] @ v[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```
